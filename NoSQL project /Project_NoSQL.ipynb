### Loading data to mangoDB

```python
import pymongo
import pprint
```

```python
import pandas as pd
from pymongo import MongoClient
import json

def mongoimport(csv_path, db_name, coll_name, db_url='localhost', db_port=27017):
    client = MongoClient(db_url, db_port)
    db = client[db_name]
    coll = db[coll_name]
    data = pd.read_csv(csv_path, encoding='latin1')
    payload = json.loads(data.to_json(orient='records'))
    coll.remove()
    coll.insert(payload)
    return coll.count()
```

## Soccer data

### 1. Use JOINS to denormalize tables

```python
import psycopg2
conn = psycopg2.connect(database='postgres',user='postgres', password='')
df1=pd.read_sql('''
SELECT A.*, B."Nationality_Name", C."Club",C."Club_Position", C."Club_Joining"
FROM
  (
    (SELECT * FROM soccer_player) A
    FULL OUTER JOIN
    (SELECT * FROM soccer_nationality) B on A."National_id" = B."National_id"
    FULL OUTER JOIN
    (SELECT * FROM soccer_club) C on A."index" = C."index"
  );''', con=conn)
```

### Q5:What posts are like mine?

We ues regex to find similar tweets. In the first example, we can see that all of the tweets contain "Lionel Messi", and all of the tweets in the second example contain "Cristiano Ronaldo".

```python
match = {
    'tweets':{"$regex": "Lionel Messi"}
}

group = {
    '_id':"$tweets"
}

ret2 = db.soccer_twitter.aggregate([
    {'$match':match},
    {'$group': group}
])

list(ret2)
```

```python
match = {
    'tweets':{"$regex": "Cristiano Ronaldo"}
}

group = {
    '_id':"$tweets"
}

ret2 = db.soccer_twitter.aggregate([
    {'$match':match},
    {'$group': group}
])

list(ret2)
```
