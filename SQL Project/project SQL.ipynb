```python
#create a table called hashtag_synonyms to store hashtag synonmous data
%sql CREATE TABLE hashtag_synonyms(tag varchar, synonmous varchar)
```

```python
#function, insert hashtag and corresponding synonyms to hashtag_synonyms table
def write_sql(tag, synonmous):
    db= psycopg2.connect(database="postgres", user="postgres", password="")
    cursor = db.cursor()
    insert_query = "INSERT INTO hashtag_synonyms(tag, synonmous) VALUES ( %s,%s)"
    cursor.execute(insert_query, (tag, synonmous))
    db.commit()
    cursor.close()
    db.close()
    return
```

```python
#get synonyms of hashtags and write data to database
for hashtag in x:
    y = wn.synsets(hashtag)
    for syn in y:
        for l in syn.lemmas():
            write_sql(hashtag, l.name())
            print(hashtag, y)
#     for t in y:
#         c = wn.synsets(t)
```

### 5. create Semantic Information table

```python
##create a table to store hashtag_semantic_category data, including 2 fields, hashtag and category
%sql create table hashtag_semantic_category(tag varchar, category varchar)
```

```python
#function, insert getting data to database
def write_sql2(tag,category):
    db= psycopg2.connect(database="postgres", user="postgres", password="")
    cursor = db.cursor()
    insert_query = "INSERT INTO hashtag_semantic_category(tag,category) VALUES (%s,%s)"
    cursor.execute(insert_query, (tag,category))
    db.commit()
    cursor.close()
    db.close()
    return
```

```python
#loop, get hashtag category data by using nltk library and load it to table hashtag_semantic_category 
for hashtag in x:
    y = wn.synsets(hashtag)
    for syn in y:
        for l in syn.topic_domains():
            for n in l.lemmas():
                write_sql2(hashtag, n.name())
                print(hashtag, n.name())
```

```python
#delete repeat rows, select distinct data to new table, drop old one
%sql select distinct * into hashtag_semantic_categories from hashtag_semantic_category
```

```python
%sql DROP TABLE hashtag_semantic_category
```

```python
#show hashtag category data from new table
%sql SELECT * FROM hashtag_semantic_categories
```

### 8.What keywords/ hashtags should I add to my post?

By counting the sum of retweet count of the same tag and ordering it, we can know that the post contains which tags is popular in a specific domain. These tags are what we would like to add to our posts.

```python
%sql SELECT soccer_hashtags.tag, sum(soccer_tweets.retweet_count) as s \
FROM soccer_tweets JOIN soccer_hashtags ON soccer_hashtags.id = soccer_tweets.id \
WHERE soccer_tweets.player_name = 'Cristiano Ronaldo' \
GROUP by soccer_hashtags.tag \
ORDER BY s DESC;
```

```python
%sql SELECT soccer_hashtags.tag, sum(soccer_tweets.retweet_count) as a \
FROM soccer_tweets JOIN soccer_hashtags ON soccer_hashtags.id = soccer_tweets.id \
WHERE soccer_tweets.player_name = 'Lionel Messi' \
GROUP by soccer_hashtags.tag \
ORDER BY a DESC;
```
